```python
import spacy
nlp = spacy.load("en_core_web_sm")
import en_core_web_sm
nlp = en_core_web_sm.load()
doc = nlp("This is a sentence.")
print([(w.text, w.pos_) for w in doc])
```

```python
for token in doc:
  print(token)
```

```python
for sent in doc.sents:
  print(sent)
```
