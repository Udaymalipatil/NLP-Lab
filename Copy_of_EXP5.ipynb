```python
import numpy as np
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import accuracy_score, classification_report

# 1. Load and preprocess the dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize pixel values (0–255) to (0–1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encode target labels
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)

# 2. Define feed-forward neural network model
model = Sequential([
    Flatten(input_shape=(28, 28)),      # Input layer (flatten 28x28 → 784)
    Dense(128, activation='relu'),      # Hidden layer 1
    Dense(64, activation='relu'),       # Hidden layer 2
    Dense(10, activation='softmax')     # Output layer (10 digits)
])

# 3. Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 4. Train the model
model.fit(X_train, y_train_cat, epochs=5, batch_size=128, verbose=1)

# 5. Evaluate the model
test_loss, test_acc = model.evaluate(X_test, y_test_cat, verbose=0)
print("\nFeed-Forward ANN Test Accuracy:", test_acc)

# 6. Display predictions for sample data
y_pred = np.argmax(model.predict(X_test), axis=1)

print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
